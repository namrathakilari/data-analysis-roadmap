```python
df=df.drop(columns=df.columns[0])
df
```

```python
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)
df
```

```python
df[["Dept","State"]]=df["Department"].str.split('_',expand=True)
df=df.drop(columns='Department')
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df["Date"]=df["Date"].dt.date
df
```

```python
df["Contact"]=df["Contact"].str.title()
df["Client"]=df["Client"].str.title()
df
```

```python
df['XNYS/XNAS'] = df['Client'].str.extract(r'\(([^)]+)\)')
df['Client'] = df['Client'].str.replace(r'\s*\([^)]+\)', '', regex=True)
df
```

```python
df["XNYS/XNAS"] = df["XNYS/XNAS"].astype(str)
df["XNAS/XNYS"] = df["XNYS/XNAS"].apply(lambda x: x.split(":", 1)[1].strip() if ":" in x else None)
df=df.drop(columns="XNYS/XNAS")
df
```
